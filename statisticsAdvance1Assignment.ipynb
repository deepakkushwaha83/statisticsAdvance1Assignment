#### Q-13.What is the Central Limit Theorem (CLT), and why is it critical in statistics?
Ans.It states that if we take a large number of random samples from any population and calculate their means,the distribution of those sample means will approach a normal distribution as the sample size increases.

Conditions of Central Limit Theorem-> 
i) The numbers of sample should be large
ii) The sample size should be greater than or equal to 30.

*Importances of CLT:-*

i)Enables Normal Approximation: Even if the population is non-normal, CLT ensures that the sampling distribution of the mean is approximately normal.

ii)Simplifies Analysis: Knowing that the sample mean follows a normal distribution allows the use of z-scores and probability calculations without needing the original distribution.

iii)Used in Real-World Applications: It is widely used in quality control, election polling, finance, and machine learning to make statistical inferences.

#### Q-21.What are the assumptions for applying the Central Limit Theorem?
Ans. The Central Limit Theorem (CLT) states that the sampling distribution of the sample mean (or sum) approaches a normal distribution as the sample size increases, regardless of the population’s original distribution. However, for CLT to hold, certain assumptions must be met:
i) Random Sampling:-	Data should be randomly selected.
ii) Independence:-	Each observation should be independent.
iii) Sample Size:-	n≥30 for skewed data, smaller n for normal data.
iv) Finite Variance:-	The population should have a finite variance.
v) Identically Distributed:-	The samples should follow the same distribution.

```python
# Q-2. Generate a discrete uniform distribution using Python and plot the probability mass function (PMF).

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Define the range of values (e.g., rolling a fair die from 1 to 6)
low, high = 1, 7 

# Generate a discrete uniform distribution
x = np.arange(low, high)  # Possible values
pmf_values = randint.pmf(x, low, high)  # Calculate PMF

# Plot the PMF
plt.bar(x, pmf_values, color='blue', alpha=0.6, edgecolor='black')

# Labels and title
plt.xlabel("Value (x)")
plt.ylabel("Probability P(X=x)")
plt.title("Discrete Uniform Distribution (PMF)")
plt.xticks(x)  # Ensure x-axis shows all values

# Show the plot
plt.show()
```

```python
# Q-3.  Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

from scipy.stats import bernoulli

def bernoulli_pdf(x, p):
    """
    Calculate the probability mass function (PMF) for a Bernoulli distribution.

    Parameters:
    x (int): The value of the random variable (0 or 1)
    p (float): Probability of success (0 ≤ p ≤ 1)

    Returns:
    float: Probability P(X = x)
    """
    if x not in [0, 1]:
        raise ValueError("x must be either 0 or 1")
    if not (0 <= p <= 1):
        raise ValueError("p must be between 0 and 1")
    
    return bernoulli.pmf(x, p)

# Example Usage:
p = 0.7  # Probability of success
print("P(X=0):", bernoulli_pdf(0, p))  # Failure probability
print("P(X=1):", bernoulli_pdf(1, p))  # Success probability
```

```python
# Q-6. Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Define the range of values (e.g., rolling a fair die from 1 to 6)
low, high = 1, 7  # randint uses an exclusive upper bound, so 7 means values [1,2,3,4,5,6]

# Generate the x values (possible outcomes)
x = np.arange(low - 1, high + 1)  # Extending for step-like plot

# Calculate CDF values
cdf_values = randint.cdf(x, low, high)

# Plot the CDF
plt.step(x, cdf_values, where='post', color='blue', label="CDF", linewidth=2)

# Labels and title
plt.xlabel("Value (x)")
plt.ylabel("Cumulative Probability P(X ≤ x)")
plt.title("Cumulative Distribution Function (CDF) of Discrete Uniform Distribution")
plt.xticks(range(low - 1, high + 1))  # Ensure all values are shown
plt.yticks(np.linspace(0, 1, high - low + 2))  # CDF ranges from 0 to 1
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()

# Show the plot
plt.show()
```

```python
#-9. Write a Python function to calculate Z-scores from a dataset and plot them.

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def calculate_z_scores(data):
    """
    Compute Z-scores for a given dataset.

    Parameters:
    data (array-like): Input dataset.

    Returns:
    np.array: Z-scores of the dataset.
    """
    mean = np.mean(data)
    std_dev = np.std(data, ddof=1)  # Using sample standard deviation
    z_scores = (data - mean) / std_dev
    return z_scores

def plot_z_scores(data):
    """
    Compute and plot the Z-scores of a dataset.

    Parameters:
    data (array-like): Input dataset.
    """
    z_scores = calculate_z_scores(data)

    # Plot histogram of Z-scores
    plt.figure(figsize=(8, 5))
    sns.histplot(z_scores, bins=30, kde=True, color="blue", edgecolor="black")
    
    # Add mean and standard deviation lines
    plt.axvline(0, color='red', linestyle='dashed', linewidth=2, label="Mean (Z=0)")
    plt.axvline(-1, color='green', linestyle='dashed', linewidth=1, label="Z=-1")
    plt.axvline(1, color='green', linestyle='dashed', linewidth=1, label="Z=1")

    # Labels and title
    plt.xlabel("Z-score")
    plt.ylabel("Frequency")
    plt.title("Histogram of Z-scores")
    plt.legend()
    plt.grid(True, linestyle="--", alpha=0.6)

    # Show the plot
    plt.show()

# Generate a sample dataset (Normally distributed)
np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=15, size=1000)

# Calculate Z-scores and plot them
plot_z_scores(sample_data)
```

```python
# Q-10.  Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Generate a non-normal distribution (Exponential)
np.random.seed(42)
population = np.random.exponential(scale=2, size=10000)  # Exponential distribution

# Function to demonstrate CLT
def central_limit_theorem(population, sample_size, num_samples):
    sample_means = []
    
    for _ in range(num_samples):
        sample = np.random.choice(population, size=sample_size, replace=True)
        sample_means.append(np.mean(sample))
    
    return np.array(sample_means)

# Different sample sizes to test CLT
sample_sizes = [5, 30, 100]  
num_samples = 1000  # Number of times we sample

# Plot the distributions
plt.figure(figsize=(12, 5))

for i, sample_size in enumerate(sample_sizes, 1):
    sample_means = central_limit_theorem(population, sample_size, num_samples)
    
    plt.subplot(1, 3, i)
    sns.histplot(sample_means, bins=30, kde=True, color="blue", edgecolor="black")
    plt.title(f"Sample Size = {sample_size}")
    plt.xlabel("Sample Mean")
    plt.ylabel("Frequency")

plt.suptitle("Central Limit Theorem (CLT) in Action")
plt.tight_layout()
plt.show()
```

```python
# Q-14.Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(data, x):
    mean = np.mean(data)
    std_dev = np.std(data, ddof=1)  # Sample standard deviation
    z_score = (x - mean) / std_dev
    return z_score

# Generate a random dataset (normal distribution)
np.random.seed(42)  # For reproducibility
data = np.random.normal(loc=50, scale=10, size=1000)  # Mean=50, Std=10, Size=1000

# Data point to calculate Z-score
x_value = 65
z_score = calculate_z_score(data, x_value)
print(f"Z-score for {x_value}: {z_score:.2f}")

# Plot standard normal distribution
x = np.linspace(-4, 4, 100)
y = norm.pdf(x, 0, 1)  # Standard normal distribution

plt.plot(x, y, color='blue', label="Standard Normal Distribution")
plt.axvline(z_score, color='red', linestyle="dashed", label=f"Z-score = {z_score:.2f}")
plt.xlabel("Z-score")
plt.ylabel("Probability Density")
plt.title("Z-score in Standard Normal Distribution")
plt.legend()
plt.show()
```

```python
# Q-15. Implement hypothesis testing using Z-statistics for a sample dataset.

import numpy as np
import scipy.stats as stats

# Given data
population_mean = 500  # Claimed mean (H0)
population_std = 10  # Known population standard deviation
sample_size = 50
sample_mean = 495  # Sample mean
alpha = 0.05  # Significance level

# Calculate Z-score
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Calculate p-value for a two-tailed test
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

# Print results
print(f"Z-score: {z_score:.2f}")
print(f"P-value: {p_value:.4f}")

# Hypothesis testing decision
if p_value < alpha:
    print("Reject the null hypothesis (H0).")
    print("There is significant evidence that the mean weight is different from 500g.")
else:
    print("Fail to reject the null hypothesis (H0).")
    print("There is no significant evidence to say the mean weight is different from 500g.")
```

```python
# Q-16.Create a confidence interval for a dataset using Python and interpret the result.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Generate a random sample dataset (normal distribution)
np.random.seed(42)
sample = np.random.normal(loc=100, scale=15, size=50)  # Mean=100, Std=15, Size=50

# Calculate sample statistics
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)  # Sample standard deviation
n = len(sample)
confidence_level = 0.95  # 95% confidence level
t_value = stats.t.ppf(1 - (1 - confidence_level) / 2, df=n-1)  # t-score for 95% confidence

# Calculate confidence interval
margin_of_error = t_value * (sample_std / np.sqrt(n))
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

# Print results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
print(f"We are {confidence_level * 100}% confident that the true population mean lies between {lower_bound:.2f} and {upper_bound:.2f}.")

# Visualizing Confidence Interval
plt.figure(figsize=(8, 4))
plt.errorbar(x=1, y=sample_mean, yerr=margin_of_error, fmt='o', color='b', capsize=5, label="95% Confidence Interval")
plt.axhline(y=lower_bound, color='r', linestyle='--', label="Lower Bound")
plt.axhline(y=upper_bound, color='g', linestyle='--', label="Upper Bound")
plt.axhline(y=sample_mean, color='b', linestyle='-', label="Sample Mean")
plt.xticks([])
plt.ylabel("Mean Value")
plt.legend()
plt.title("Confidence Interval for Sample Mean")
plt.show()
```

```python
# Q-17. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Step 1: Generate data from a normal distribution
np.random.seed(42)
population_mean = 50  # True mean
population_std = 10   # True standard deviation
sample_size = 50      # Sample size
sample = np.random.normal(loc=population_mean, scale=population_std, size=sample_size)

# Step 2: Calculate sample statistics
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)  # Sample standard deviation (Bessel's correction)
confidence_level = 0.95  # 95% confidence level
t_value = stats.t.ppf(1 - (1 - confidence_level) / 2, df=sample_size - 1)  # t-score

# Step 3: Compute confidence interval
margin_of_error = t_value * (sample_std / np.sqrt(sample_size))
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

# Step 4: Print and Interpret Results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
print(f"We are {confidence_level * 100}% confident that the true population mean lies between {lower_bound:.2f} and {upper_bound:.2f}.")

# Step 5: Visualize the Confidence Interval
plt.figure(figsize=(8, 4))
plt.errorbar(x=1, y=sample_mean, yerr=margin_of_error, fmt='o', color='b', capsize=5, label="95% Confidence Interval")
plt.axhline(y=lower_bound, color='r', linestyle='--', label="Lower Bound")
plt.axhline(y=upper_bound, color='g', linestyle='--', label="Upper Bound")
plt.axhline(y=sample_mean, color='b', linestyle='-', label="Sample Mean")
plt.axhline(y=population_mean, color='purple', linestyle=':', label="True Mean")
plt.xticks([])
plt.ylabel("Mean Value")
plt.legend()
plt.title("Confidence Interval for Sample Mean")
plt.show()
```

```python
# Q-18. Write a Python script to calculate and visualize the probability density function (PDF) of a normal distribution.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define parameters for the normal distribution
mean = 0      # Mean (μ)
std_dev = 1   # Standard Deviation (σ)

# Generate x values (range covering most of the distribution)
x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)

# Compute the Probability Density Function (PDF)
pdf_values = norm.pdf(x, mean, std_dev)

# Plot the PDF
plt.figure(figsize=(8, 5))
plt.plot(x, pdf_values, color='b', label=f"Normal Distribution (μ={mean}, σ={std_dev})")
plt.xlabel("X values")
plt.ylabel("Probability Density")
plt.title("Probability Density Function (PDF) of Normal Distribution")
plt.axvline(mean, color='r', linestyle="--", label="Mean (μ)")
plt.legend()
plt.grid()
plt.show()
```

```python
# Q-19. Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution.


import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Define the Poisson distribution parameter (lambda)
lambda_value = 4  # Average occurrences per interval

# Generate x values (possible number of events)
x_values = np.arange(0, 15, 1)  # Events from 0 to 14

# Calculate the Cumulative Distribution Function (CDF)
cdf_values = poisson.cdf(x_values, lambda_value)

# Plot the CDF
plt.figure(figsize=(8, 5))
plt.step(x_values, cdf_values, color='b', where="mid", label=f"Poisson(λ={lambda_value})")
plt.xlabel("Number of Events (k)")
plt.ylabel("Cumulative Probability")
plt.title("Cumulative Distribution Function (CDF) of a Poisson Distribution")
plt.axhline(y=0.95, color='r', linestyle="--", label="95% Probability Threshold")
plt.legend()
plt.grid()
plt.show()

# Example interpretation
k_value = 6
probability = poisson.cdf(k_value, lambda_value)
print(f"P(X ≤ {k_value}) = {probability:.4f}")
print(f"There is a {probability * 100:.2f}% chance that 6 or fewer events will occur in the given interval.")

```

```python
# Q-20. Simulate a random variable using a continuous uniform distribution and calculate its expected value.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

# Define the range [a, b] of the uniform distribution
a, b = 10, 20  # Lower and upper bounds

# Generate a random sample of size 1000
np.random.seed(42)
sample_size = 1000
random_sample = np.random.uniform(a, b, sample_size)

# Calculate the expected value (theoretical mean)
expected_value = (a + b) / 2

# Calculate the sample mean (approximation of expected value)
sample_mean = np.mean(random_sample)

# Plot histogram of the sampled data
plt.figure(figsize=(8, 5))
plt.hist(random_sample, bins=20, density=True, alpha=0.6, color='b', label="Sampled Data")

# Plot the theoretical uniform distribution PDF
x = np.linspace(a, b, 100)
pdf_values = uniform.pdf(x, a, b-a)
plt.plot(x, pdf_values, 'r-', label="Uniform PDF")

# Add vertical lines for expected value and sample mean
plt.axvline(expected_value, color='g', linestyle="--", label=f"Expected Value (E[X]) = {expected_value}")
plt.axvline(sample_mean, color='purple', linestyle="--", label=f"Sample Mean = {sample_mean:.2f}")

plt.xlabel("X values")
plt.ylabel("Probability Density")
plt.title("Continuous Uniform Distribution & Expected Value")
plt.legend()
plt.grid()
plt.show()

# Print results
print(f"Theoretical Expected Value: {expected_value}")
print(f"Sample Mean (from simulation): {sample_mean:.2f}")
```

```python
# Q-21.Write a Python program to compare the standard deviations of two datasets and visualize the difference.

import numpy as np
import matplotlib.pyplot as plt

# Generate two datasets with different standard deviations
np.random.seed(42)
dataset1 = np.random.normal(loc=50, scale=5, size=1000)   # Mean=50, Std=5
dataset2 = np.random.normal(loc=50, scale=15, size=1000)  # Mean=50, Std=15

# Calculate standard deviations
std1 = np.std(dataset1, ddof=1)  # Sample standard deviation
std2 = np.std(dataset2, ddof=1)

# Plot histograms
plt.figure(figsize=(10, 5))
plt.hist(dataset1, bins=30, alpha=0.6, color='b', label=f"Dataset 1 (σ={std1:.2f})", density=True)
plt.hist(dataset2, bins=30, alpha=0.6, color='r', label=f"Dataset 2 (σ={std2:.2f})", density=True)

# Add labels and title
plt.axvline(np.mean(dataset1), color='blue', linestyle="dashed", linewidth=2, label="Mean (Dataset 1)")
plt.axvline(np.mean(dataset2), color='red', linestyle="dashed", linewidth=2, label="Mean (Dataset 2)")
plt.xlabel("Value")
plt.ylabel("Density")
plt.title("Comparison of Standard Deviations")
plt.legend()
plt.grid()
plt.show()

# Print results
print(f"Dataset 1 - Standard Deviation: {std1:.2f}")
print(f"Dataset 2 - Standard Deviation: {std2:.2f}")
print("Dataset 2 has a higher standard deviation, meaning its values are more spread out.")
```

```python
# Q-22.Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution.

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate a dataset from a normal distribution
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=1000)  # Mean=50, Std=10, Sample size=1000

# Calculate Range
data_range = np.max(data) - np.min(data)

# Calculate Interquartile Range (IQR)
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

# Print results
print(f"Range: {data_range:.2f}")
print(f"Interquartile Range (IQR): {IQR:.2f}")

# Boxplot for visualization
plt.figure(figsize=(8, 5))
sns.boxplot(data, color="skyblue")
plt.title("Boxplot of Normally Distributed Data")
plt.xlabel("Values")
plt.grid()
plt.show()
```

```python
# Q-23. Implement Z-score normalization on a dataset and visualize its transformation.

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate a dataset from a normal distribution
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=1000)  # Mean=50, Std=10, Sample size=1000

# Compute mean and standard deviation
mean = np.mean(data)
std = np.std(data, ddof=1)

# Apply Z-score normalization
z_scores = (data - mean) / std

# Plot original and normalized data distributions
plt.figure(figsize=(12, 5))

# Original data histogram
plt.subplot(1, 2, 1)
sns.histplot(data, bins=30, kde=True, color="blue")
plt.axvline(mean, color='r', linestyle="--", label="Mean")
plt.title("Original Data Distribution")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.legend()

# Z-score normalized data histogram
plt.subplot(1, 2, 2)
sns.histplot(z_scores, bins=30, kde=True, color="green")
plt.axvline(0, color='r', linestyle="--", label="Mean (0)")
plt.title("Z-score Normalized Data")
plt.xlabel("Z-score Values")
plt.ylabel("Frequency")
plt.legend()

plt.tight_layout()
plt.show()

# Print results
print(f"Original Mean: {mean:.2f}, Original Std Dev: {std:.2f}")
print(f"Normalized Data -> Mean: {np.mean(z_scores):.2f}, Std Dev: {np.std(z_scores):.2f}")
```

```python
# Q-24.Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

def analyze_skewness_kurtosis(mean=50, std_dev=10, size=1000):
    """
    Generates a dataset from a normal distribution and calculates skewness & kurtosis.
    """
    # Generate normally distributed data
    np.random.seed(42)
    data = np.random.normal(loc=mean, scale=std_dev, size=size)

    # Calculate skewness and kurtosis
    skewness = stats.skew(data)
    kurtosis = stats.kurtosis(data, fisher=True)  # Fisher=True gives excess kurtosis

    # Plot histogram with KDE
    plt.figure(figsize=(8, 5))
    sns.histplot(data, bins=30, kde=True, color='skyblue')
    plt.axvline(np.mean(data), color='r', linestyle="--", label=f"Mean = {np.mean(data):.2f}")
    plt.title("Histogram of Normally Distributed Data")
    plt.xlabel("Values")
    plt.ylabel("Frequency")
    plt.legend()
    plt.grid()
    plt.show()

    # Print results
    print(f"Skewness: {skewness:.4f}")
    print(f"Kurtosis (Excess): {kurtosis:.4f}")

# Call the function
analyze_skewness_kurtosis()
```
